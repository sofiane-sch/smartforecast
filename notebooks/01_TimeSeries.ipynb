```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go
from ipywidgets import interact
```

```python
def linear_trend(x:np.array, coeff:float=1, intercept:float=0):
    return coeff*x + intercept

def exponential_trend(x:np.array, coeff:float=1, intercept:float=0):
    return np.exp(x) + intercept - 1

def seasonality(x:np.array, freq:float=1, phi:float=0):
    return np.sin(2*np.pi*freq*x + phi)
```

```python
np.arange(2,10,1)
```

```python
time = np.arange(101)
```

```python
fig = go.Figure()

# Add traces, one for each slider step
for frequency in np.arange(0, 0.05, 0.001):
    fig.add_trace(
        go.Scatter(
            visible=False,
            line=dict(color="#00CED1", width=6),
            name="ùúà = " + str(frequency),
            x=time,
            y=seasonality(time, frequency)))

# Make 10th trace visible
fig.data[10].visible = True

# Create and add slider
steps = []
for i in range(len(fig.data)):
    step = dict(
        method="update",
        args=[{"visible": [False] * len(fig.data)},
              {"title": "Slider switched to step: " + str(i)}],  # layout attribute
    )
    step["args"][0]["visible"][i] = True  # Toggle i'th trace to "visible"
    steps.append(step)

sliders = [dict(
    active=10,
    currentvalue={"prefix": "Frequency: "},
    pad={"t": 50},
    steps=steps
)]

fig.update_layout(
    sliders=sliders
)

fig.show()
```
